```python
import pandas as pd

df= pd.read_csv('C:/Users/lenovoku/DEEP LEARNING/praktikum_1.csv')
df
```

```python
pip install -U scikit-learn
```

```python
from sklearn.model_selection import train_test_split

X = df.drop('y', axis=1).values
Y = df['y'].values

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42) 
```

```python
X_train, x_val, Y_train, y_val = train_test_split(x_train, y_train, test_size=0.2, random_state=42) 
```

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(16, input_dim=4, activation='tanh'))
model.add(Dense(8, activation='tanh'))
model.add(Dense(4, activation='tanh'))
model.add(Dense(2, activation='tanh'))
model.add(Dense(1, activation='tanh'))
```

```python
model.compile(loss='mse', optimizer='adam')
model.summary()
```

# Training Model

```python
catatan = model.fit(X_train, Y_train, validation_data=[x_val, y_val], epochs=100, batch_size=64)
```

# PLOT PERFORMA MODEL

```python
import matplotlib.pyplot as plt

plt.plot(range(len(catatan.history['loss'])), catatan.history['loss'], 'r',
        range(len(catatan.history['val_loss'])), catatan.history['val_loss'], 'y')
plt.show()

#JIKA JARAK LOSS LEBIH BESAR JAUH DARIPADA VALIDATION LOSS DISEBUT 'OVERFITTING'
#JIKA JARAK LOSS LEBIH KECIL DARIPADA VALIDATION LOSS DISEBUT 'UNDERFITTING'
```

# EVALUASI MODEL

```python
loss = model.evaluate(x_test, y_test)
print('Loss Model = ', loss)
```
